```python
#create helper fun
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
mov['genres'] = mov['genres'].apply(convert)
```

```python
#create helper func for cast
import ast
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
mov['cast'] = mov['cast'].apply(convert3)
```

```python
mov['crew'][0]
#job=director --- name
```

```python
#create helper fun
import ast
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
mov['crew'] = mov['crew'].apply(fetch_director)
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#funcr
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return ' '.join(y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
# now we have vectors , shape is (4608,5000)
# so all the 5000 vectors are drawn - cal distance b/w all vectors
# here we cal distance
#1. Eclidean distance: we'll not caling this dist becuase: this is dist b/w tip of vectors
#2. cosine dist : instead we cal angel b/w them
        # if angel is small then it's same vector
# Eclidean dist is fail in High Dimensional data--use cosine dist

# dist is inversily prop to similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    dist = similarity[movie_index]
    movies_list = sorted(list(enumerate(dist)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
